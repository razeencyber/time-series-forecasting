```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

### Data Prep

```python
df = pd.read_csv('PJME_hourly.csv')
df.head()
```

```python
df.tail()
```

```python
# Check for missing values

df.isnull().sum()
```

```python
df = df.set_index('Datetime')
df.index = pd.to_datetime(df.index)
```

```python

df.plot(style='.',
        figsize=(10, 4),
        color=color_pal[2],
        title='Energy (MW)')

plt.grid(axis='y', linestyle='', which='both')

plt.show()
```

### Training and Testing

```python
train = df.loc[df.index < '01-01-2015']
test = df.loc[df.index >= '01-01-2015']

fig, ax = plt.subplots(figsize=(15, 5))
train.plot(ax=ax, label='Training Set', color = '#C0C0C0', title='Data Train/Test Split')
test.plot(ax=ax, label='Test Set', color = '#FF5349')
ax.axvline('01-01-2015', color='black', ls='--')

# Removing the grid lines 

ax.grid(axis='x', linestyle='', which='both')  


ax.legend(['Training Set', 'Test Set'])
plt.show()
```

```python
week_data = df.loc[(df.index > '01-01-2010') & (df.index < '01-08-2010')]
week_data.plot(figsize=(15, 5), title='Week Worth of Data') 
plt.show()
```

```python
def features_creation(df):
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    df['weekofyear'] = df.index.isocalendar().week
    return df

df = features_creation(df)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.boxplot(data=df, x='hour', y='PJME_MW')
ax.set_title('Energy Consumption (MW) by Hour')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.boxplot(data=df, x='month', y='PJME_MW', palette='Blues')
ax.set_title('Energy Consumption by Month')
plt.show()
```

### Model Creation

```python
train = features_creation(train)
test = features_creation(test)

FEATURES = ['dayofyear', 'hour', 'dayofweek', 'quarter', 'month', 'year']
TARGET = 'PJME_MW'

X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

### Implementing XG Boost


```python
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

### Cross Validation

```python
X_train = np.arange(len(train)).reshape(-1, 1)  
y_train = train['PJME_MW'].values

tscv = TimeSeriesSplit(n_splits=5)
cv_results = []

for train_idx, val_idx in tscv.split(X_train):
    X_t, X_v = X_train[train_idx], X_train[val_idx]
    y_t, y_v = y_train[train_idx], y_train[val_idx]
    
    model = LinearRegression()
    model.fit(X_t, y_t)
    
    y_pred = model.predict(X_v)
    rmse = np.sqrt(mean_squared_error(y_v, y_pred))
    cv_results.append(rmse)

cv_results
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', color = "#87CEEB", title='Feature Importance')
plt.show()
```

```python
test['prediction'] = reg.predict(X_test)

if 'prediction' in df.columns:
    df = df.drop(columns=['prediction'])

df = df.merge(test[['prediction']], how='left', left_index=True, right_index=True)


ax = df[['PJME_MW']].plot(figsize=(15, 5), color='#C0C0C0')  
df['prediction'].plot(ax=ax, style='.', color='#FF5349')  

plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Data and Prediction')
plt.show()

```
